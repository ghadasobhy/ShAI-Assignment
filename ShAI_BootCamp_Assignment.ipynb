# Data Cleaning

```python
#Handle missing data 
# drop nan data
data_new = df.dropna(axis=1, how= 'any' , subset=None, inplace=False)
data_new
```

```python
print(data_new.isnull().sum())
```

```python
data_new.head(10)
```

# Basic Data Visualization

```python
#histograms 
import matplotlib.pyplot as plt
plt.hist(data_new['TotalPay'], bins=20)
# Add labels and title
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.title('Distribution of TotalPay')
# Show the plot
plt.show()
```

```python
JobTitle = data_new['JobTitle']
JobTitle.head(10)
```

```python
# Count the number of employees in each department
department_counts = data_new['JobTitle'].value_counts()
# Create the pie chart
plt.pie(department_counts, startangle=90 ,autopct='%1.1f%%' )
plt.title("Employee Distribution by Department")
plt.axis("equal")  # Equal aspect ratio ensures a circular chart
plt.show()
```

# Grouped Analysis

```python
# Grouping a DataFrame by Multiple Columns
groupedby = data_new.groupby(['JobTitle', 'Agency','Year'])
groupedby
```

```python
# Calculate the Sum for Each Group
print(data_new.groupby(['JobTitle', 'Agency','Year'])['TotalPay'].sum())
```

```python
#summary statistics for each group
average_salary_by_group = groupedby['TotalPay'].mean()
print(average_salary_by_group)
```

```python
average_salary_by_group = groupedby['TotalPay'].describe()
print(average_salary_by_group)
```

# Simple Correlation Analysis

```python
# Calculate Year correlation
Year_corr = data_new['TotalPay'].corr(data_new['Year'])
Year_corr
```

```python
# Create the scatter plot
plt.scatter(data_new['Year'],data_new['TotalPay'] , color='b', marker='o', label='TotalPay vs. Year')
plt.xlabel('Year')
plt.ylabel('TotalPay')
plt.title('TotalPay vs. Year')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate TotalPayBenefits correlation
TotalPayBenefits_corr = data_new['TotalPay'].corr(data_new['TotalPayBenefits'])
TotalPayBenefits_corr
```

```python
# Create the scatter plot
plt.scatter(data_new['TotalPayBenefits'],data_new['TotalPay'] , color='b', marker='o', label='TotalPay vs. TotalPayBenefits')
plt.xlabel('TotalPayBenefits')
plt.ylabel('TotalPay')
plt.title('TotalPay vs. TotalPayBenefits')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter(data_new['Id'],data_new['TotalPay'] , color='b', marker='o', label='TotalPay vs. Id')
plt.xlabel('Id')
plt.ylabel('TotalPay')
plt.title('TotalPay vs. Id')
plt.grid(True)
plt.legend()
plt.show()
```
